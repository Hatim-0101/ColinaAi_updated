```python
import numpy as np
import cv2
import os

import cv2
import numpy as np


def load_video(path, frame_count=40, target_size=(299, 299)):
    cap = cv2.VideoCapture(path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = np.linspace(0, total_frames - 1, frame_count, dtype=int)

    for frame_index in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
        ret, frame = cap.read()
        if ret:
            frame = cv2.resize(frame, target_size)
            frame = frame / 255.0
            frames.append(frame)
        else:
            break

    cap.release()
    video_array = np.array(frames)

    return video_array


def create_dataset(directory, frame_count=40, target_size=(299, 299)):
    videos = []
    labels = []

    for folder in ["nofoul", "foul"]:
        folder_path = os.path.join(directory, folder)
        for filename in os.listdir(folder_path):
            video_path = os.path.join(folder_path, filename)
            video = load_video(video_path, frame_count=frame_count, target_size=target_size)
            if video.shape[0] == frame_count:
                videos.append(video)
                label = 1 if folder == "foul" else 0
                labels.append(label)

    return np.array(videos), np.array(labels)
```

```python
videos, labels = create_dataset("C:/Users/iturk/OneDrive/Desktop/Dataset_org")
```

```python
X_train, X_val, y_train, y_val = train_test_split(videos, labels, test_size=0.15, random_state=42)
```

```python
K.clear_session()
def inc_block(inputs):
    x0 = Conv3D(64,
                (1,1,1),
                activation='relu',
                padding="same")(inputs)
    x1 = Conv3D(64,
                (1, 1, 1),
                activation='relu',
                padding="same")(inputs)
    x1 = Conv3D(64,
                (3, 3, 3),
                activation='relu',
                padding="same")(x1)
    x2 = Conv3D(64,
                (1, 1, 1),
                activation='relu',
                padding="same")(inputs)
    x2 = Conv3D(64,
                (3, 3, 3),
                activation='relu',
                padding="same")(x2)
    x3 = MaxPooling3D((3, 3, 3),
                      (1,1,1),
                      padding="same")(inputs)
    x3 = Conv3D(64,
                (1,1,1),
                activation='relu',
                padding="same")(x3)
    outputs = tf.concat([x0, x1, x2, x3], 4)
    return outputs
```

```python
def I3D(inputs):
  x = Conv3D(64,
             (7,7,7),
             (2,2,2),
             activation='relu')(inputs)
  x = MaxPooling3D((3,3,3),
                   (2,2,2),
                   padding='same')(x)
  x = Conv3D(64,
             (1,1,1),
             activation='relu')(x)
  x = Conv3D(64,
             (3,3,3),
             activation='relu')(x)
  x = MaxPooling3D((3,3,3),
                   (2,2,2),
                   padding='same')(x)
  x = inc_block(x)
  x = inc_block(x)
  x = MaxPooling3D((3,3,3),
                   (2,2,2),
                   padding='same')(x)
  x = inc_block(x)
  x = inc_block(x)
  x = inc_block(x)
  x = inc_block(x)
  x = inc_block(x)
  x = MaxPooling3D((2,2,2),
                   (2,2,2),
                   padding='same')(x)
  x = inc_block(x)
  x = inc_block(x)
  x = AvgPool3D((2,7,7),
                strides=(1,1,1),
                padding='same')(x)
  x = Conv3D(64,
             (1,1,1))(x)
  x = Flatten()(x)
  x = Dense(128,
            activation='relu')(x)
  # x = Dense(1,
  #           activation='sigmoid')(x)
  return x
```

```python
# the poese data and segment data must be stacked before enter the i3d netwark!!
```

```python
inputsfirst = Input(shape=(40, 299, 299, 3))
outputsfirst = I3D(inputsfirst)
```

```python
inputsPoese = Input(shape=(40,299, 299, 3)) # must be stacked
outputpoese = I3D(inputsPoese)
```

```python
inputsSegment = Input(shape=(40,299, 299, 3)) # must be stacked
outputsSegment = I3D(inputsSegment)
```

```python
combined_output = tf.add(outputpoese, outputsSegment)# addition +
```

```python


final_output_toPredict = tf.add(outputsfirst, combined_output) # addition +

# combined_model = Model(inputs=[inputs, inputs_i3d], outputs=final_output)
```

```python
 x = Dense(1, activation='sigmoid')(final_output_toPredict)
```

```python
model = Model(inputs=[inputsfirst , inputsPoese , inputsSegment],
              outputs=x)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
history = model.fit(X_train,
                    y_train,
                    batch_size=8,
                    validation_data = (X_val, y_val),
                    epochs=20,
                    callbacks=[tensorboard_callback])
```
